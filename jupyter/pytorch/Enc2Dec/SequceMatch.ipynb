```python
from difflib import SequenceMatcher
```

```python
src = open("./data/dev.src").readlines()
trg = open("./data/dev.trg").readlines()
```

```python
assert len(src) == len(trg)
```

```python
fw = open("./data/dev_mask.txt","w")

for i in range(len(src)):
    source = src[i].strip().split()
    target = trg[i].strip().split()
    
    matcher = SequenceMatcher(None, source, target)
    
    fin_source = []
    diffs = list(matcher.get_opcodes())
    for diff in diffs:
        tag,i1,i2,j1,j2 = diff
#         print(diff)
        if tag == 'equal':
            fin_source.extend(source[i1:i2])
        else:
            fin_source.append('<s>')
            fin_source.extend(source[i1:i2])
            fin_source.append('<\s>')
    print("source is:{}\n target is:{}\n fin_source is:{}\n".format(" ".join(source), " ".join(target)," ".join(fin_source)))
#     print(" ".join(fin_source))
    fw.write(" ".join(fin_source) + "\n")
        
    
```
