```python
#torch.norm()是对输入的tensor求范数
#torch.norm(input, p)，其中p指定范数，默认为2，即求2-范数
#下面看一个例子

import torch

a = torch.ones((2,3))
print("a is:",a)
a1 = torch.norm(a, p=1)#指定求1范数
print("1范数是：", a1)
a2 = torch.norm(a)#默认求2范数
print("2范数是：", a2)

```

```python
#求指定维度上的范数
#torch.norm(input, p, dim, out=None, keepdim=False) -> Tensor
#返回输入张量给定维dim上的每行的p范数
#参数：
#   Input(Tensor) -- 输入张量
#   p(float) -- 范数计算中的幂指数值
#   dim(int) -- 缩减的维度，也就是指定计算的维度
#   out(Tensor, optional) -- 结果张量
#   keepdim(bool)  -- 保持输出的维度（此参数官方文档中未给出，但是很常用）

#下面我们看一个dim的例子
import torch

a = torch.tensor([[1,2,3,4],[1,2,3,4]]).float() #norm仅支持floatTensor，a是一个2*4的Tensor
#a = torch.tensor([[1,2,3,4],[1,2,3,4]])会报错，因为torch.tensor()创建的是LongTensor类型的数据
#a = torch.Tensor([[1,2,3,4],[1,2,3,4]])#torch.Tensor()不会报错，因为他创建的就是floatTensor
a0 = torch.norm(a, p=2, dim=0) #在0维度上求2范数
a1 = torch.norm(a, p=2, dim=1) #在1维度上求2范数
print("p2 result when dim=0:", a0)
print("p2 result when dim=1:",a1)



#再看一个keepdim的例子，其含义就是保持输出的维度，很抽象，我们还是通过具体例子来看看
a = torch.rand((2,3,4))
a_true = torch.norm(a, p=2, dim=1, keepdim=True)#保持维度
a_false = torch.norm(a, p=2, dim=1, keepdim=False)#不保持维度
a_default = torch.norm(a, p=2, dim=1) #默认keepdim为false
print("the original shape of a is:", a.size())
print("the shape when keepdim is true:", a_true.size())
print("the shape when keepdim is false:", a_false.size())
print("the shape when keepdim is default:", a_default.size())

```
