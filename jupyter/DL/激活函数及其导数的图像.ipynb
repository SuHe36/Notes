本文是对四个常用的激活函数及其导数的图像的表示--sigmoid、tanh、Relu、leaky Relu；



# sigmoid函数

sigmoid是常用的激活函数之一，其计算公式为：
$$sigmoid(x) = \frac{1}{1+e^{-x}}  $$
    
sigmoid的导数为：
$$sigmoid(x) = \frac{e^{-x}}{(1 + e^{-x})^2}  = sigmoid(x)(1-sigmoid(x))$$

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(-10, 10, 0.01)
y = 1/(1+np.exp(-x))

#sigmoid的函数图像
plt.plot(x,y)
plt.grid(color='b', linewidth='0.1', linestyle='--')

#sigmoid的导数图像
dev_y = y*(1-y)
plt.plot(x,dev_y)
plt.grid(color='b', linewidth='0.2', linestyle='--')

```

# tanh函数

tanh函数又称为双曲正切函数，其计算公式为：
$$tanh(x) = \frac{sinh(x)}{cosh(x)} = \frac{e^z - e^{-z}}{e^z + e^{-z}}  $$

tanh的导数为：
$$tanh(x)' = 1 - tanh(x)^2 $$



```python

x = np.arange(-10, 10, 0.01)
y = (np.exp(x) - np.exp(-x))/(np.exp(x) + np.exp(-x))

#tanh(x)的图像
plt.plot(x,y)
plt.grid(color='b', linewidth='0.1', linestyle='--')

#tanh(x)的导数的图像
dev_y = 1-np.square(y)
plt.plot(x, dev_y)
plt.grid(color='b', linewidth='0.1', linestyle='--')




```

# Relu函数

Relu是一个分段函数，其计算公式为：
$$
f(x) = \begin{cases}
x, &x>=0 \\ 0, &x<0
\end{cases}
$$

Relu的导数为：
$$
f(x)' = \begin{cases}
1, &x>=0 \\ 0, &x<0
\end{cases}
$$



```python
x = np.arange(-10, 10, 0.01)
y = list(map(lambda x:x if x>0 else 0, x))

#Relu的图像
plt.plot(x,y)
plt.grid(color='b', linewidth='0.1', linestyle='--')

#Relu的导数的图像
dev_y = list(map(lambda x:1 if x>0 else 0, x))
plt.plot(x, dev_y)
plt.grid(color='b', linewidth='0.1',linestyle='--')

```

# leaky Relu函数

leaky Relu是Relu的一种变形，其计算公式为：
$$
f(x) = \begin{cases}
x,& if x>=0 \\ \lambda x, & if x<0
\end{cases}
$$

其中$\lambda$是取值范围为(0,1)之间的数
leaky Relu的导数的图像为：
$$
f(x)' = \begin{cases}
1, & if x>=0 \\ \lambda ,& if x <0
\end{cases}
$$


```python
x = np.arange(-10, 10, 0.01)
lambda1 = 0.1 #指定系数lambda为0.1
y = list(map(lambda x:x if x>=0 else lambda1*x, x))


#leaky Relu的函数图像
plt.plot(x,y)
plt.grid(color='b', linewidth='0.1', linestyle='--')

#leaky Relu的导数的图像
dev_y = list(map(lambda x:1 if x>=0 else lambda1, x))
plt.plot(x, dev_y)
plt.grid(color='b', linewidth='0.1', linestyle='--')


```
